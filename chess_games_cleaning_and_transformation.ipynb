# Deleting redundant columns

```python
data.columns
```

```python
columns_to_drop = ['id','created_at', 'last_move_at', 'increment_code' , 'white_id', 'black_id']
columns_to_keep = ['rated', 'turns', 'victory_status',
       'winner','white_rating','black_rating', 'moves', 'opening_eco', 'opening_name','opening_ply']
```

```python
len(columns_to_keep)
```

```python
len(columns_to_drop)
```

# 1st Method, Filtering Data

```python
df = data[columns_to_keep]
```

```python
df
```

# 2nd Method, Dropping Columns

```python
data.drop(columns = columns_to_drop, inplace = True)
```

```python
data.shape
```

# Renaming the columns

```python
data.columns
```

```python
data.rename(columns = {"turns": "Number_of_turns", "moves": "moves_in_chess_notation", "opening_ply": "number_of_moves_in_opening"}, inplace = True)
```

```python
new_column_names = []
for i in data.columns:
    new_column_names.append(i.lower())
```
